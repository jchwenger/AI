Each transformation is performed by a computational 'layer'. The number of layers is the model depth. Models with three or more layers - successive transformations are 'deep'. So there we have it: deep learning models perform three or more successive transformations. 

Deep models are almost invariably **neural networks**

Deep models are almost invariably neural networks. In other words, a statistically deep model is implemented - as software - by a neural network. But what is a neural network?

Although neural networks were developed by analogy to the function of the brain, they do not emulate brain function

They do not 'think' or learn like brains

A neural network is a stacked sequence of computational layers 

Neural networks, or artificial neural networks to give them their proper title, were originally crude models of actual, biological, neural networks, i.e. brains. Brains are fantastically complex networks of intercommunicating 'neurons'. Although neural networks were developed by analogy to the function of the brain, they do not emulate brain function. In particular, they do not 'think' or learn like brains - yet. We regard neural networks rather more prosaically: as a stacked sequence of computational layers.

Here are six examples. Each small 28 x 28 pixel image bears a handwritten digit. Neural networks work like this:

The 28 x 28 images are 'vectorised' by placing each pixel value in a vector - a 1D array of 784 elements. The bar coded strip represents the vectorised data. Each bar is a non zero pixel value; the darker the bar, the larger the value.

The sample is presented to the first layer of the network. Layers are drawn as thin boxes containing 'neuronal' blobs. Traditional depictions of neural networks show individual connections between these blobs - officially known as 'units' - but this picture is unnecessary. Instead, arrows indicate the forward data flow.

The data is transformed by the layer to a 512 element vector. Notice that the transformed representation is smaller than the original - from 784 elements to 512 - and less sparse - many vector elements are non-zero.

The 512 element vector is fed into a second layer.

A second transformation produces a ten element vector of floating point values. Each element represents a classification of the input: '0', '1', ‘2’...; the values quantify confidence in the network prediction. In this case the output vector has a large value in the eight slot - the network is fairly confident in its prediction: the image is a ‘7’ (the first element of the output vector corresponds to the digit ‘0’, the second as ‘1’ and so on).
