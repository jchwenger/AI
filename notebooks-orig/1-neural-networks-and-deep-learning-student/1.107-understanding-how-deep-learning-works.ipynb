The layer transformation is governed by a number of adjustable parameters - 'weights' and 'biases'

You can think of the weights and biases as a large number of sliders - like a huge mixer desk. Every possible mixer desk configuration corresponds to a particular transformation.

\begin{align*}
x' &= w_{11} x + w_{12} y \\
y' &= w_{21} x + w_{22} y
\end{align*}

Before, in the red/black example, the first layer had four parameters - four sliders - the w’s.

Neural networks are a sequence of simple data transformations - parameterised by weights and biases - mapping input samples to output predictions.

A two layer network, showing parameter mixer desks, the input sample and the output prediction.

Network performance is evaluated with a 'loss function' that compares prediction to target. The loss function is arbitrary  - up to some restrictions. The loss has to be zero for a perfect match - when prediction is identical to target - and otherwise positive. There are other mathematical constraints - more later. 

Weights and biases are initially set at random values and the loss is likely to be very high. An optimiser makes small changes in order to reduce loss. The loss slowly drops over many training loops and the network becomes a better predictor. The optimiser advantageously implements a backwards layer by layer correction know as ‘backpropagation’. Samples are fed forward through the network, losses calculated, and parameter values are tweaked layer by layer, working from right to left. The entire data set is processed over many training loops and the loss gradually diminishes until the network makes pretty reliable predictions. The network is then tested on unseen data.
