Step 3: we build our neural network.

```python
# imports
from tensorflow.keras import models, layers
```

Import `models` and `layers`

```python
# an empty network
network = models.Sequential()
```

First create an empty network object. Sequential means the network is a series of transformational layers. Data moves in a single `forward` direction through the network - a `feed-forward` neural network.

Here is the planned two layer network architecture.

```python
# add two layers
# don't worry about the complicated arguments - they will be explained later
network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28, )))
network.add(layers.Dense(10, activation='softmax'))
```

And here is the code. `network.add()` adds a layer. Each layer has a few specific attributes tyat are specified in the call to `layers.Dense`. We will investigate layer attributes shortly: just try to follow the main flow of the program for now.

The second **softmax** layer outputs a vector whose elements form a **probability distribution**

- the numbers are nonnegative and sum to one - a probability distribution

- outputs are interpreted as probabilities of membership of each class: the probability that the input sample is labeled '0' or '1' or '2' and so on

The second layer is a `softmax layer` meaning that it outputs a vector of nonnegative floats that sum to one, that is, a probability distribution. The elements of the softmax output vector are interpreted as probabilities of membership of each class: the probability that the input sample is labeled '0' or '1' or '2' and so on.

We must specify a loss function, an optimiser and one or more training metrics.

```python
# optimiser, loss and metrics are chosen at compilation
network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy', 
                metrics=['accuracy'])
```

The training specification is made at 'compilation'.

`rmsprop` is one of several `tensorflow.keras` optimizers

Categorical crossentropy is the preferred loss for single label multiclass problems

`accuracy` is the fraction of correctly classified samples 

`rmsprop` is one of several `tensorflow.keras` optimizers. Categorical crossentropy is the preferred loss for  problems in which each sample belongs to one of several classes. Once more, don't worry what thsi piece of jargon means. `accuracy` is a simple metric: the fraction of correctly classified samples.
