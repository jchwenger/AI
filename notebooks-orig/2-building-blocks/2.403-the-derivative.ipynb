A smooth function $f$ is a function without abrupt changes. This means that $f$ in the vicinity of a point $x$ can always be approximated $f(x + \delta x) \approx f(x) + m \delta x$ where $m$ is the gradient of the tangent at $x$ and $\delta x$ is a small step.

The gradient of differentiable functions - those without abrupt changes - is calculable from simple rules. The gradient function of $f = x^n$, $f'(\text{also written }\frac{df}{dx})$ is $nx^{n-1}$.

For example, the gradient function of $f(x) = x^2$ is $f'(x) = 2x$

For example, the gradient function of $f(x) = x^2$ is $f'(x) = 2x$. The gradient at $x = 3$ is $6$.
