### Plotting activations

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

#### Helpers

```python
n = 5
x = tf.Variable(tf.linspace(-n, n, 100))

def plot_activation(x, activation_fns):
    fig, (axes) = plt.subplots(1, 2, figsize=(10, 4))

    for fn in activation_fns:

        # using GradienTape to compute the gradient
        with tf.GradientTape() as tape:
            y = fn(x)

        # plot the forward pass
        axes[0].plot(x.numpy(), y.numpy(), label=fn.__name__)
        
        # retrieve and plot the gradient (backward pass)
        dy = tape.gradient(y, x)        
        axes[1].plot(x.numpy(), dy.numpy(), label=fn.__name__)

    descrs = ["Forward pass", "Backward pass (gradient)"]
    # first plot: forward
    axes[0].set_xlim(-n, n)
    axes[0].set_ylim(-n, n)
    axes[0].grid()
    axes[0].set_title(descrs[0])
    axes[0].legend()
    
    # second plot: backward
    axes[1].grid()
    axes[1].set_title(descrs[1])
    axes[1].legend()

    plt.show()
```

#### ReLu, LeakyReLu

```python
activations_list = [
    # max(0, x)
    tf.keras.activations.relu,
    # can take an arg, `negative_slope`, that controls the slope when x < 0
    tf.keras.activations.leaky_relu,
]
plot_activation(x, activations_list)
```

#### Tanh

```python
activations_list = [
    # sinh(x) / cosh(x), i.e. tanh(x) = ((exp(x) - exp(-x)) / (exp(x) + exp(-x)))
    tf.keras.activations.tanh,
    # x / (abs(x) + 1), easier to compute than tanh
    tf.keras.activations.softsign,
]
plot_activation(x, activations_list)
```

#### More!

[activations doc](https://www.tensorflow.org/api_docs/python/tf/keras/activations)

```python
activations_list = [
    # alpha * exp(x) - 1 if x < 0 else x, alpha=1.0
    tf.keras.activations.elu,
    # scale * alpha * (exp(x) - 1) if x < 0 else scale * x if x > 0 (alpha=1.67326324 scale=1.05070098)
    tf.keras.activations.selu,
    # x * P(X <= x), i.e. 0.5 * x * (1 + erf(x / sqrt(2)))
    tf.keras.activations.gelu,
]

plot_activation(x, activations_list)
```

```python
activations_list = [
    # log(exp(x) + 1)  
    tf.keras.activations.softplus,
    # x * tanh(softplus(x))
    tf.keras.activations.mish,
    # note that in TF swish is considered the same as silu (x * sigmoid(x))
    # whereas swish allows for a beta parameters: x * sigmoid(beta * x) – same as silu for beta = 1
    tf.keras.activations.silu
]

plot_activation(x, activations_list)
```
